```python
import numpy as np
BLACK = -1
WHITE = 1
BAR = 0;
```

```python
def roll_dice():
    return np.random.randint(1,7,2)
```

```python
def game_over(board):
    return board.W_free == 15 or board.B_free
```

```python
import numpy as np

B = -1
W = 1
BAR = 0
class Board:
    def __init__(self):
        self.board = np.zeros(25, dtype=int)
    
        self.board[1] = 2*B
        self.board[6] = 5*W
        self.board[8] = 3*W
        self.board[12] = 5*B
        self.board[13] = 5*W
        self.board[17] = 3*B
        self.board[19] = 5*B
        self.board[24] = 2*W
        
        self.B_bar = 0
        self.W_bar = 0
        
        self.B_free = 0
        self.W_free = 0
        
        self.B_home = 5
        self.W_home = 5
        
        
    
    def update(self):
        self.B_home = sum(self.board[19:25][self.board[19:25]<0])
        self.W_home = sum(self.board[1:7][self.board[1:7]>0])
        
        
    def move(self,start_point, die, player):
        
        if player == W:
            end_point = start_point - die
            if start_point == BAR:
                self.board[25-die] += 1
                self.W_bar += -1
                
            elif end_point < 1:
                if self.W_free + self.W_home == 15:
                    self.board[start_point] += -1
                    self.W_free +=1 
                
            else:
                if self.board[end_point] == -1:
                    self.board[end_point] += 1
                    self.B_bar += 1
                self.board[start_point] += -1
                self.board[end_point] += 1
                
        elif player == B:
            end_point = start_point + die
            if start_point == BAR:
                self.board[0+die] += 1
                self.B_bar += -1
                
            elif end_point > 24:
                if self.B_free + self.B_home == 15:
                    self.board[start_point] += 1
                    self.B_free +=1 
                
            else:
                if self.board[end_point] == 1:
                    self.board[end_point] += -1
                    self.W_bar += 1
                self.board[start_point] += +1
                self.board[end_point] += -1
        update()
                
    def print(self):
#       W = o
#       B = x
        print('__________________________________________________________', end = '')
        print('__________________________________________________________________')
        print('__________________________________________________________', end = '')
        print('________________________________________o HOME BOARD______________')
        print('| |\t12\t11\t10\t9\t8\t7\t| |\t6\t5\t4\t3\t2\t1\t| | o-FREE:')
        
        for j in range(1,max(abs(self.board))+1): 
            print('| |',end = '')
            for i in range(12,6,-1):
                if self.board[i] >= j:
                    print('\to', end = '')
                elif self.board[i] <= -j:
                    print('\tx', end = '')
                else:
                    print(' \t', end = '')
            print('\t| |', end='')

            for i in range(6,0,-1):
                if self.board[i] >= j:
                    print('\to', end = '')
                elif self.board[i] <= -j:
                    print('\tx', end = '')
                else:
                    print(' \t', end = '')
    
            
            if j != 1 and j != max(abs(self.board))-1 and j != max(abs(self.board)):
                print('\t| |')
            
            if j == 1:
                print('\t| |', end = '')
                print('\t', self.W_free)
                
            if j == max(abs(self.board))-1:
                print('\t| |', end = '')
                print('  o-BAR:')
                
            if j == max(abs(self.board)):
                print('\t| |', end = '')
                print('\t', self.W_bar)
    
        print('| |',end = '')
        print('                                                     ', end = '')
        print('| |',end = '')
        print('                                                     ', end = '')
        print('| |')
        print('----------------------------------------------------------', end = '')
        print('---------------------------------------------------------')
        print('| |',end = '')
        print('                                                     ', end = '')
        print('| |',end = '')
        print('                                                     ', end = '')
        print('| |')
        
        for j in range(max(abs(self.board)),0,-1): 
            print('| |',end = '')
            for i in range(13,19):
                if self.board[i] >= j:
                    print('\to', end = '')
                elif self.board[i] <= -j:
                    print('\tx', end = '')
                else:
                    print(' \t', end = '')
            print('\t| |', end='')
            for i in range(19,25):
                if self.board[i] >= j:
                    print('\to', end = '')
                elif self.board[i] <= -j:
                    print('\tx', end = '')
                else:
                    print(' \t', end = '')
            
            if j != 1 and j != max(abs(self.board))-1 and j != max(abs(self.board)):
                print('\t| |')
            
            if j == 1:
                print('\t| |', end = '')
                print(' x-FREE:')
                
            if j == max(abs(self.board))-1:
                print('\t| |', end = '')
                print('\t', self.W_bar)
                
            if j == max(abs(self.board)):
                print('\t| |', end = '')
                print('  x-BAR:')
            
        print('| |\t13\t14\t15\t16\t17\t18\t| |\t19\t20\t21\t22\t23\t24\t| |\t', self.B_bar)
        
        print('__________________________________________________________', end = '')
        print('__________________________________________________________________')
        print('__________________________________________________________', end = '')
        print('________________________________________x HOME BOARD______________')

        
        return
```
