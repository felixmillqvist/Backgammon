```python
def roll_dice():
    dices = np.random.randint(1,7,2)
    if dices[0] == dices[1]:
        return np.array([dices[0], dices[1], dices[1], dices[1]])
    return dices
```

```python
class Player:
    BLACK = -1
    WHITE = 1
    BAR = 0
    def __init__(self, color):
        self.color = color
        self.bar = 0
        self.free = 0
        self.home = 5
        self.moves = []
        self.dices = []
        
    def bearing_off_possible(self):
        return self.home + self.free == 15
    
    def has_dead_piece(self):
        return self.free > 0
    
    def increase_free(self):
        self.free += 1
        
    def decrease_free(self):
        self.free += -1
        
    def increase_bar(self):
        self.bar += 1
        
    def decrease_bar(self):
        self.bar += -1
        
    def increase_home(self):
        self.home += 1
    
    def decrease_home(self):
        self.home += -1
        
    def is_done(self):
        return self.free == 15
    
    def possible_moves(self, board):
        self.moves = []
        moves = []
        for die in self.dices:
            
            if self.color == WHITE:
                if self.has_dead_piece() and board[25-die] > -2:
                    moves.append([BAR, die])
                
                else:
                    if self.bearing_off_possible():
                        if board[die] > 0:
                            moves.append([die,die])
                        
                        elif not self.is_done():
                            start_pip = np.max(np.where(board[1:7]>0))+1
                            if start_pip < die:
                                moves.append([start_pip, die])
                    
                    for start_pip in np.where(board[1:25]>0)[0]+1:
                        end_pip = start_pip-die
                        if end_pip > 0 and board[end_pip] > -2:
                            moves.append([start_pip,die])
                        
            elif self.color == BLACK:
                
                if self.has_dead_piece() and board[0+die] > 2:
                    moves.append([BAR, die])
                
                else:
                    if self.bearing_off_possible():
                        if board[25-die] < 0:
                            moves.append([25-die,die])
                        
                        elif not self.is_done():
                            start_pip = np.min(np.where(board[19:25]<0))+19
                            if 25-start_pip < die:
                                moves.append([start_pip, die])
                    
                    for start_pip in np.where(board[1:25]<0)[0]+1:
                        end_pip = start_pip+die
                        if end_pip < 25 and board[end_pip] < 2:
                            moves.append([start_pip,die])
               
        self.moves = moves    
        return moves
    
    
    
        
```
